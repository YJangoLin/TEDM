```python
import pandas as pd 
import numpy as np
```

```python
yoloResData = pd.read_csv("./data/results.txt", sep=r'\s+', header=None)
yoloResData.head()
```

```python
columns = ['训练轮数', 'GPU消耗', 'train/box_loss', 'train/obj_loss ','train/cls_loss','train/total_loss','本轮中数据中目标数量','输入图片大小','P','R','mAP@.5','mAP@.5:.95','val/Box_loss','val/obj_loss','val/cls_loss']

yoloResData.columns = columns
yoloResData.head()
```

```python
anaData = yoloResData[["训练轮数", "P", "R", "mAP@.5", "mAP@.5:.95"]]
anaData.head()
```

```python
import matplotlib.pyplot as plt
plt.plot(anaData.index, anaData['P'], label="precsion")
plt.plot(anaData.index, anaData['R'], label="recall")
plt.xlabel('epoch')
plt.legend()
plt.show()
```
